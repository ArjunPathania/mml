## 🔹 1. Dot Product

```python
# TODO: Implement dot product manually
def dot_manual(a, b):
    return  sum(a[i]*b[i] for i in range(len(a)))# sum of product of individual elements 
#necessary condition len(a)==len(b)

# Test
a = [1,2,3]
b = [4,5,6]
print(dot_manual(a,b))
```

## 🔹 2. Vector Norm

```python
# TODO: Implement norm manually
def norm_manual_l2(v):
    return sum(v[i]**2 for i in range(len(v))) # l2 norm

def norm_manual_l1(v):
    return sum(abs(v[i]) for i in range(len(v))) 

print(norm_manual_l1([3,4,12,-3]))
print(norm_manual_l2([-2,-4,-5]))
```

## 🔹 3. Projection

```python
# TODO: Implement projection
def projection(v, u):
    num_=sum (v[i]*u[i] for i in range(len(v)))
    denom_=sum(u[i]**2 for i in range(len(u)))
    res = [float(num_/denom_ * u[i]) for i in range(len(u))]
    return res

v = np.array([3,4])
u = np.array([1,0])
print(projection(v,u))
```

```python
def projection(v, u):
    if len(v) != len(u):
        raise ValueError("Vectors must have same dimension")

    denom = sum(x*x for x in u)
    if denom == 0:
        raise ValueError("Cannot project onto zero vector")

    num = sum(a*b for a, b in zip(v, u))

    return [(num / denom) * x for x in u]

v = np.array([3,4])
u = np.array([1,0])
print(projection(v,u))
```

```python
# NumPy reference
def projection_np(v,u):
    return (np.dot(v,u)/np.dot(u,u))*u

print(projection_np(v,u))
```

## 🔹 4. Eigenvalues & Eigenvectors

```python
# Use NumPy first
A = np.array([[2,0],[0,3]])
vals, vecs = np.linalg.eig(A)
print(vals)
print(vecs)
```

# TODO:

```python
Av = A@vecs[0]
λv = vals*vecs[0]
print(Av)
print(λv)
print(Av==λv)
```

## 🔹 5. SVD

```python
A = np.array([[1,2],[3,4]])
U,S,Vt = np.linalg.svd(A)
print(U)
print(S)
print(Vt)
```

# TODO:
- Reconstruct matrix using U, S, Vt

```python
# Reconstruction
S_mat = np.zeros((U.shape[0], Vt.shape[0]))
np.fill_diagonal(S_mat, S)
A_recon = U @ S_mat @ Vt
print(A_recon)
```

## 🔹 6. Low Rank Approximation

```python
# TODO: keep only top k singular values
k = 1
S_k = np.zeros((U.shape[0], Vt.shape[0]))
S_k[:k,:k] = np.diag(S[:k])
A_k = U @ S_k @ Vt
print(A_k)
```
